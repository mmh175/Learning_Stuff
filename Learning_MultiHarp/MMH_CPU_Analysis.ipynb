#### Importing necessary libraries

```python

import os, time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pyptu import PTUParser
from joblib import Parallel, delayed, parallel_backend
from tqdm import tqdm
```

#### Selecting input and output directories

```python

PC          = os.getlogin()
base_dir    = fr'C:\Users\{PC}\Documents\JupyterNotebook\Summer_25\UTC\TQN610'
template    = "Channel_2_3"
in_file     = os.path.join(base_dir, f"{template}.ptu")
out_dir     = os.path.join(base_dir, template)  
os.makedirs(out_dir, exist_ok=True)
```

#### Making _chunk files_

Only writes _Channel_ and _Time Tag_ to save time and space.

```python

start_0   = time.time()
# ────────────────────────── Load and chunk ──────────────────────────

parser    = PTUParser(in_file)
parser.load()
df        = parser.photons[["Channel", "Time Tag"]]
n_events  = len(df)
N_chunks  = 10
chunk_sz  = -(-n_events // N_chunks)

def save_chunk(idx: int):
    start = idx * chunk_sz
    end   = min((idx + 1) * chunk_sz, n_events)
    out   = os.path.join(out_dir, f"chunk_{idx + 1:02d}.txt")
    df.iloc[start:end].to_csv(out, sep="\t", index=False)

Parallel(n_jobs=-1)( delayed(save_chunk)(i) for i in tqdm(range(N_chunks), desc="Writing chunks") )

stop_0 = time.time()
# ─────────────────────────────────────────────────────────────────────
print(f'TIme: {stop_0 - start_0} sec')
```

#### Analyzing the data

Initial conditions:

- Channel 2 is heralding channel, channel 3 is heralded (signal) channel
- Scanning from 4.4M - 4.8M bins (22us - 24us), in 100 bins (500 ps) step size.
   

```python

start_1 = time.time()
# ─────────────────────────────────────────────────────────────────────

df = pd.read_csv(
    fr'{base_dir}\{template}\chunk_01.txt',
    sep    ='\t'
)
df.columns = df.columns.str.strip() 

heralds    = np.sort(np.asarray(df.loc[df['Channel'] == 2, 'Time Tag'].values))
signals    = np.sort(np.asarray(df.loc[df['Channel'] == 3, 'Time Tag'].values))

delays     = np.arange(4_400_000, 4_800_001, 1_00)

def count_for_delay(d):
    a      = heralds + d
    b      = a + 100
    lefts  = np.searchsorted(signals, a, side='left')
    rights = np.searchsorted(signals, b, side='left')
    counts = rights - lefts
    return int(np.sum(counts))  

with parallel_backend('threading', n_jobs=-1):
    counts = Parallel()(
        delayed(count_for_delay)(d) for d in tqdm(delays, desc="Processing Delays")
    )

# ─────────────────────────────────────────────────────────────────────
stop_1 = time.time()

print(f"Total time taken: {stop_1 - start_1:.2f} seconds")
```

#### Writing the data into a .csv file

```python

start_2 = time.time()
# ─────────────────────────────────────────────────────────────────────

output_df = pd.DataFrame({
    'Delay': delays.asnumpy() if hasattr(delays, 'asnumpy') else delays,
    'Counts': counts
})

output_path = fr'{base_dir}\{template}\output_counts.csv'
output_df.to_csv(output_path, index=False)
print(f"Results saved to {output_path}")

# ─────────────────────────────────────────────────────────────────────
stop_2 = time.time()

print(f"Total time taken: {stop_2 - start_2:.2f} seconds")
```
